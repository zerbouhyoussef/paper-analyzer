```python
import pytesseract
from pdf2image import convert_from_path
import string
import time 
```

```python
def pdf_to_text(pdf_path):
    #Convert PDF to images 
    images = convert_from_path(pdf_path)
    text = ""
    for image in images:
        text += pytesseract.image_to_string(image)
    return text
```

```python
time_start = time.time()
extracted_text = pdf_to_text("data/Changing Data Sources in the Age of Machine Learning for Official Statistics.pdf")
time_end = time.time()
```

```python
def clean_text(text):
    #Remove punctuation and newlines
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator).replace('\n', ' ')
```

```python
def alphabetic_ratio(text):
    if not text:
        return 0.0

    alpha_count = sum(c.isalpha() for c in text)
    total_count = len(text)

    return alpha_count / total_count
```

```python
length_extracted = len(extracted_text)
cleaned_text = clean_text(extracted_text)
length_cleaned = len(cleaned_text)
ratio = alphabetic_ratio(cleaned_text)
```

```python
print(f"Time taken: {time_end - time_start:.2f} seconds")
print(f"Length of extracted text: {length_extracted}")
print(f"Length of cleaned text: {length_cleaned}")
print(f"Alphabetic character ratio: {ratio:.2%}")
```
